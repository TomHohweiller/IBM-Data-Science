```python
import numpy as np
import matplotlib.pyplot as plt

def x1(t):
    return 2 + 2*np.cos(t)
def y1(t):
    return 2 + 2*np.sin(t)
def x2(t):
    return 2 - np.cos(t/2)
def y2(t):
    return 2 - np.sin(t/2)
def x3(t):
    return 1 + np.cos(t)/10
def y3(t):
    return 3 + np.sin(t)/10
def x4(t):
    return 3 + np.cos(t)/10
def y4(t):
    return 3 + np.sin(t)/10

t = np.linspace(0, 2 * np.pi, 200)

plt.plot(x1(t), y1(t))
plt.plot(x2(t), y2(t))
plt.plot(x3(t), y3(t))
plt.plot(x4(t), y4(t))
plt.legend({'Hello there'})

plt.show()
```
